```python
## Web Scrapped the Traveller Type Data ####
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

url_template = "https://www.airlinequality.com/airline-reviews/delta-air-lines/page/{}/?sortby=post_date%3ADesc&pagesize=100"

traveler_types = []
for i in range(1, 30):
    url = url_template.format(i)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    traveller_type_tags = soup.find_all("td", class_='review-rating-header type_of_traveller') 
    for tag in traveller_type_tags:
        traveller_type = tag.find_next_sibling('td').text.strip()
        traveler_types.append(traveller_type)

print(traveler_types)
```

```python
## Create other lists to store important features for building a comprehensive dataset for sentiment analysis
#create ann empty list to collect all reviews
reviews  = []

#create an empty list to collect rating stars
stars = []

#create an empty list to collect country the reviewer is from
country = []

#create an empty list to collect date the reviewer is from
date = []

# soup.find_all is a method int he BeautifulSoup library to search for all occurances of a particular HTML ta
# The soup variable refers to a BeautifulSoup object, which is an object that represents the parsed HTML content of a webpage. 
# Please note lxml is the newest edition of web scrapping parser in python
# it is compatible with complex HTML and XML
# lxml is faster and feature-rich   

from bs4 import BeautifulSoup
import requests 

for i in range(1, 36):
    page = requests.get(f"https://www.airlinequality.com/airline-reviews/delta-air-lines/page/{i}/?sortby=post_date%3ADesc&pagesize=100")
    
    soup = BeautifulSoup(page.content, "lxml")

# soup.find_all is a method int he BeautifulSoup library to search for all occurances of a particular HTML ta
# The soup variable refers to a BeautifulSoup object, which is an object that represents the parsed HTML content of a webpage. 
# Please note lxml is the newest edition of web scrapping parser in python
# it is compatible with complex HTML and XML
# lxml is faster and feature-rich
    
    for item in soup.find_all("div", class_="text_content"):
        reviews.append(item.text)

# div item prop, class="rating-10"
    for item in soup.find_all("div", class_ = "rating-10"):
        try:
            stars.append(item.span.text)
        except:
            print(f"Error on page {i}")
            stars.append("None")
            
    #date
    for item in soup.find_all("time"):
        date.append(item.text)
        
    #country
    for item in soup.find_all("h3"):
        country.append(item.span.next_sibling.text.strip(" ()"))
```

```python
# We also need to scrap recommendation results for sentiment analysis
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# The base URL of the page to scrape
url_base = "https://www.airlinequality.com/airline-reviews/delta-air-lines/?sortby=post_date%3ADesc&pagesize=100"

# Create an empty list to store the recommendation values
recommendations = []

# Loop through all the pages of reviews
for page in range(1, 34):  # There are 33 pages of reviews
    # Create the URL for the current page of reviews
    url = url_base.format(page)

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Parse the HTML content of the page with Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all the table cells with the class "review-value"
    recommendation_cells = soup.find_all("td", {"class": "review-value"})

    # Extract the recommendation values from the cells
    for cell in recommendation_cells:
        # Check if the text of the cell is "yes" or "no" and add it to the recommendations list accordingly
        recommendation = cell.get_text(strip=True)
        if recommendation == "yes":
            recommendations.append("Yes")
        elif recommendation == "no":
            recommendations.append("No")

# Print the recommendation values
print(recommendations)
```
