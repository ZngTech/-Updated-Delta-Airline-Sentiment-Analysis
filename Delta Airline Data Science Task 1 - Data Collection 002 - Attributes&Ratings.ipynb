```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# The base URL of the page to scrape
url_base = "https://www.airlinequality.com/airline-reviews/delta-air-lines/page/{}/?sortby=post_date%3ADesc&pagesize=100"

# Create an empty list to store the review attributes
attributes = []

# Create an empty list to store the rating values
ratings = []

# Loop through all the pages of reviews
page = 1
while True:
    # Create the URL for the current page of reviews
    url = url_base.format(page)

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Parse the HTML content of the page with Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all the reviews on the page
    reviews = soup.find_all("article", {"itemprop": "review"})

    # If there are no more reviews on the page, break out of the loop
    if not reviews:
        break

    # Loop through the reviews and extract the header attributes and ratings
    for review in reviews:
        headers = review.find_all("td", {"class": "review-rating-header"})
        for header in headers:
            header_text = header.text.strip()
            if header_text not in ["Recommended", "Aircraft", "Type of Traveller", "Route","Date Flown","Seat Type"]:
                attributes.append(header_text)

        rating_cells = review.find_all("td", {"class": "review-rating-stars stars"})
        for cell in rating_cells:
            rating_spans = cell.find_all("span", {"class": "star fill"})
            rating = len(rating_spans)
            ratings.append(rating)

    # Increment the page number for the next iteration
    page += 1

# Exclude the first 5 elements in the ratings list
ratings = ratings[5:]

# Print the list of review attributes and ratings
print(attributes)
print(ratings)
```
